```python
import numpy as np
from scipy import stats
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def ks_2samp_vaildation(mu1: float, sigma1: float, mu2: float, sigma2: float):
    np.random.seed(12345678) #fix random seed to get the same result
    n1 = 200  # size of first sample
    n2 = 200  # size of second sample
    rvs1 = stats.norm.rvs(size=n1, loc=mu1, scale=sigma1)
    rvs2 = stats.norm.rvs(size=n2, loc=mu2, scale=sigma2)
    
    plt.plot(rvs1, stats.norm.pdf(rvs1, mu1, sigma1), color='g',linewidth = 3)
    plt.plot(rvs2, stats.norm.pdf(rvs2, mu2, sigma2), color='r',linewidth = 3)
    return stats.ks_2samp(rvs1, rvs2)
```

```python
if __name__=="__main__":
    mu1 = 6.66317749
    sigma1 = 0.828458
    
    mu2 = 6.66317749
    sigma2 = 0.8284588632125883
    k2, p =  ks_2samp_vaildation(mu1, sigma1, mu2, sigma2)
    print("probability is : {}".format(p))
    alpha = 1e-3
    if p < alpha:
        print("Not belong to one distribution hypothesis")
    else:
        print("Belong to one distribution")
```
