```python
import numpy as np
from scipy import stats
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def example_method():
    np.random.seed(12345678) #fix random seed to get the same result
    n1 = 200  # size of first sample
    n2 = 300  # size of second sample
    rvs1 = stats.norm.rvs(size=n1, loc=0., scale=1.)
    rvs2 = stats.norm.rvs(size=n2, loc=258.89, scale=100)
    rvs3 = stats.norm.rvs(size=n2, loc=258.89, scale=255.666)
    
    plt.plot(rvs2, stats.norm.pdf(rvs2, 258.89, 100), color='g',linewidth = 3)
    plt.plot(rvs3, stats.norm.pdf(rvs3, 258.89, 255.666), color='r',linewidth = 3)
    print(stats.ks_2samp(rvs2, rvs3))
```

```python
if __name__=="__main__":
    example_method()
```
